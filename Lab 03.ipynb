````python
def unrate(outlist):

  orating = rating[rating["film_id"].isin(outlist)]

  orating = orating.groupby("film_id")[["rating"]].mean()

  return orating

unrate(output)
````

````python
def outfilm(title, limit):

  movie_list = movies[movies['title'].str.contains(title)]

  movie_indices = movie_list.iloc[0]['film_id']

  movie_indices = italia[italia['film_id'] == movie_indices]

  movie_indices = movie_indices.index[0]

  score, indices = model.kneighbors(csrorigin[movie_indices],n_neighbors=limit+1)

  score = score.flatten().tolist()

  indices = indices.flatten().tolist()

  output = italia.loc[indices]["film_id"]

  result = movies[movies["film_id"].isin(output)]

  result = pandas.merge(result, unrate(output), on="film_id")

  result["score"] = [round(i, 5) for i in score]

  return result

outfilm(title, 5)
````
